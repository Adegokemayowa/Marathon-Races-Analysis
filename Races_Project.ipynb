```python
#importing libraries
import pandas as pd
import seaborn as sns
```

```python
#loading the data
df = pd.read_csv('/content/drive/MyDrive/TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
#See what imported data looks like
df.head(10)
```

```python
#checking the shape of the data
df.shape
```

```python
#checking the datatype of each column
df.dtypes
```

```python
#Cleaning up the data
```

```python
#Filtering the data to show only 50km and 50-mile races in the USA in 2020
#Step 1 filtering the data to show only 50km and 50-miles races using the isin function
```

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
# filtering to include only events from 2020
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
#adding the USA filter
# using the .contains() function to filter out 'USA' from the string in the "Event Name" column
```

```python
df[df['Event name'].str.contains(r'\(USA\)', regex=True)]
```

```python
# combining all the filters together
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains(r'\(USA\)', regex=True))]
```

```python
#renaming the filtered data as a new dataframe
```

```python
df_2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains(r'\(USA\)', regex=True))]
```

```python
df_2.head()
```

```python
df_2.shape
```

```python
# removing '(USA)' from the events column as all the events are in the USA
```

```python
df_2['Event name'] = df_2['Event name'].str.split('(').str.get(0)
```

```python
#to check if the country has been properly removed
df_2.head(10)
```

```python
# Creating a new column, calculating the age of each athlete
```

```python
df_2['athlete_age'] = 2020 - df_2['Athlete year of birth']
```

```python
# removing the 'h' from the data in the 'Athlete Performance' column
```

```python
df_2['Athlete performance'] = df_2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df_2['Athlete performance']
```

```python
df_2.head()
```

```python
#dropping columns; Athlete club, Athlete Country, Athlete year of birth, Athlete Age Category
```

```python
df_2 = df_2.drop(columns=['Athlete club','Athlete country','Athlete year of birth','Athlete age category'])
```

```python
df_2.head()
```

```python
# cleaning up null values
df_2.isnull().sum()
```

```python
df_2[df_2['athlete_age'].isnull()]
```

```python
df_2 = df_2.dropna()
```

```python
df_2.shape
```

```python
# checking for duplicates
df_2.duplicated().sum()
```

```python
#resetting index
df_2.reset_index(drop=True)
```

```python
# changing the types of the columns
df_2.dtypes
```

```python
# changing the type of the atlethe age from float64 to int
df_2['athlete_age'] = df_2['athlete_age'].astype(int)
```

```python
# changing the type of the atlethe avg speed from object to float
df_2['Athlete average speed'] = df_2['Athlete average speed'].astype(float)
```

```python
df_2.dtypes
```

```python
# renaming columns
```

```python
df_2 = df_2.rename(columns={
                            'Year of event':'year',
                            'Event dates' : 'race_date',
                            'Event name':'race_name',
                             'Event distance/length':'race_distance',
                             'Event number of finishers' : 'race_number_of_finishes',
                             'Athlete performance': 'athlete_performance',
                             'Athlete gender':'athlete_gender',
                             'Athlete name':'athlete_name',
                             'Athlete average speed':'athlete_avg_speed',
                             'Athlete ID' : 'athlete_id'
                             })
```

```python
df_2.head()
```

```python
sns.histplot(df_2['race_distance'])
```

```python
sns.histplot(df_2, x = 'race_distance', hue = 'athlete_gender')
```

```python
# Differnece in speed for the 50km,50mi male to female
```

```python
df_2.groupby(['race_distance','athlete_gender'])['athlete_avg_speed'].mean()
```

```python
# top 15 age groups are the best in the 50km race (at least 20 races)
```

```python
df_2.query('race_distance == "50mi"').groupby(['athlete_age'])['athlete_avg_speed'].agg(['mean','count']).sort_values(by = 'mean', ascending = False).query('count > 19').head(15)
```

```python
# bottom 15 age groups are the best in the 50km race (at least 20 races)
df_2.query('race_distance == "50mi"').groupby(['athlete_age'])['athlete_avg_speed'].agg(['mean','count']).sort_values(by = 'mean', ascending = True).query('count > 19').head(15)
```

```python
# Checking if seasons of the year affect speed of the runners
# The months for each season:
# Spring : 3-5
# Summer : 6-8
# Fall : 9-11
# Winter 12-2
```

```python
df_2['race_month'] = df_2['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df_2['race_season'] = df_2['race_month'].apply(lambda x: 'Spring' if 3 <= x <= 5 else ('Summer' if 6 <= x <= 8 else ('Fall' if 9 <= x <= 11 else 'Winter')))
```
